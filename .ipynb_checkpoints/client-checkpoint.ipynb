```python
import socket
import cv2 as cv
import numpy as np
import datetime
```

```python
s = socket.socket()
host = socket.gethostname()
port = 12345
```

```python
s.connect((host,port))
```

```python
frame=None
c=0
while True:
    data = s.recv(1024)
    #sb = data.decode()
    #if sb=='frame_sent':
    #if frame!=None:
        #print(len(frame))
    if frame!=None and len(frame)==921600:
        y = np.frombuffer(frame,dtype='uint8')
        #print(len(y))
        y = y.reshape(480,640,3)
        print(datetime.datetime.now())
        cv.imshow('Live Web-Cam Feed', y)
        
        key = cv.waitKey(1)
        if key in [27, 81, 113]:#esc,Q,q
            break
        frame=None
    
    if frame is None:
        frame = data
    else:
        frame+=data

cv.destroyAllWindows()
s.close()
    
```
